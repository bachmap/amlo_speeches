```python
import pandas as pd
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api._errors import TranscriptsDisabled, NoTranscriptFound, VideoUnavailable

# read data
videos = pd.read_csv('amlo_videos_meta.csv')

# change items names 'id' in 'video_id' 
videos.rename(columns={'id': 'video_id'}, inplace=True)

# Kontrollieren, ob alles passt
videos.head()

# Transkript-Funktion
def get_transcript(video_id):
    try:
        # Hol das Transkript in Spanisch
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['es'])
        # Zu einem Fließtext verbinden
        full_text = ' '.join([entry['text'] for entry in transcript])

        print(f"✅ Transkript erfolgreich für Video ID: {video_id}")
        return full_text

    except (TranscriptsDisabled, NoTranscriptFound, VideoUnavailable) as e:
        print(f"❌ Kein Transkript für Video ID: {video_id} ({e})")
        return ""

    except Exception as e:
        print(f"⚠️ Fehler bei Video ID: {video_id} ({e})")
        return ""

# get transcripts
videos['transcript'] = videos['video_id'].apply(get_transcript)

# see results
videos[['upload_date', 'title', 'video_id', 'transcript']].head()

# save results
videos.to_csv('amlo_videos_with_transcripts.csv', index=False)

print("✅ Alle Transkripte sind gespeichert!")
```
